```python
df.columns

df_corr = df[['popularity','duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()
df_corr

    
```

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(df_corr, annot=True, fmt=".2f", cmap='coolwarm', ax=ax)
plt.title('Correlation Matrix of Spotify Track Features', fontsize=16)
```

```python
# count values in the 'track_genre' column
df["track_genre"].value_counts()
```

```python
df["track_genre"].value_counts()

df_genres = df[df["track_genre"].isin(['pop', 'hip hop', 'rock', 'dance', 'r&b', 'country'])]

# use existing sample_n (or available) to avoid sampling more rows than exist
sample_n = min(5000, len(df_genres))

df_sample = df_genres.sample(n=sample_n, random_state=1)

sns.pairplot(df_sample[["acousticness", "energy", "track_genre"]], hue="track_genre", height=5, aspect=1.5, diag_kind="kde")

print(df_sample["track_genre"].value_counts())
```
